Se crea el DataFrame

```python
df = pd.DataFrame(productos)
```

Cálculo del valor total del inventario por producto

```python
df['inventario_total_producto'] = df['precio'] * df['cantidad_disponible']
```

Valor total del inventario.

```python
df['inventario_total'] = df["inventario_total_producto"].cumsum()
```

Se simulan algunas ventas y se actualiza la cantidad de productos vendidos.

```python
df['cantidad_disponible'] = df['cantidad_disponible'] - 10
```

Se muestra la cantidad restante de cada producto después de las ventas.

```python
print(df[['nombre', 'cantidad_disponible']])
```

Se crea el DataFrame con las columnas: nombres(productos) y cantidad(productos)

```python
productos = {
    'Nombres': df['nombre'],
    'Cantidad de Productos': df['cantidad_disponible']
}
```

Se realiza el gráfivo de Barras, donde el eje x es nombrado como 'Nombre de los Productos', el eje y 'Cantidad Disponible' y el título del gráfico es 'Cantidad de Productos Disponibles'

```python
df = pd.DataFrame(productos)
df.plot(kind='bar', x='Nombres', y='Cantidad de Productos', color='blue')
plt.xlabel('Nombre de los Productos')
plt.ylabel('Cantidad Disponible')
plt.title('Cantidad de Productos Disponibles')
plt.show()
```
